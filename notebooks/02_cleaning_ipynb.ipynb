## HANDLE MISSING VALUES
### Purpose: Ensure no null issues in analytics

```python
df['description'] = df['description'].fillna(np.nan)
df['tags'] = df['tags'].fillna('')
```

##DROP IRRELEVANT COLUMNS
### Purpose: Remove non-analytical fields

```python
df.drop(['thumbnail_link'], axis=1, inplace=True)
```

## DATATYPE CONVERSION
### Purpose: Enable time-based analysis

```python
# Fix datetime
df['publishedAt'] = pd.to_datetime(df['publishedAt'])

df['publish_time'] = pd.to_datetime(df['publishedAt'])
df['trending_date'] = pd.to_datetime(df['trending_date'])
```

## REMOVE INVALID RECORDS
### Purpose: Ensure data quality

```python
df = df[df['view_count'] > 0]
```

## REMOVE DUPLICATES
### Purpose: Avoid duplicate daily entries

```python
df = df.drop_duplicates(subset=['video_id', 'trending_date'])
```

##FEATURE ENGINEERING
### Purpose: Create analytical metrics

```python
# Engagement metrics
df['likes_ratio'] = df['likes'] / df['view_count']
df['comments_ratio'] = df['comment_count'] / df['view_count']
```

```python
# Time-based features
df['publish_date'] = df['publish_time'].dt.date
df['publish_hour'] = df['publish_time'].dt.hour

df['trending_year'] = df['trending_date'].dt.year
df['trending_month'] = df['trending_date'].dt.month
df['trending_day'] = df['trending_date'].dt.day_name()
```

```python
# Virality metric
df['days_to_trend'] = (
    df['trending_date']-
    df['publish_time']
).dt.days
df['days_to_trend'] = df['days_to_trend'].apply(lambda x: x if x >= 0 else 0)
```

##TREND DURATION FEATURE
### Purpose: Measure how long videos stay trending

```python
trend_days = df.groupby('video_id')['trending_date'].nunique()
df['trend_duration'] = df['video_id'].map(trend_days)
```

## TAG CLEANING - BASIC NORMALIZATION
### Purpose: Standardize tag format

```python
import re

def clean_tags(tags):
    if tags == '[none]' or tags == '':
        return ''

    tags = tags.lower()  # normalize case
    tags = tags.replace('|', ' ')  # replace separator

    # remove special characters
    tags = re.sub(r'[^a-zA-Z ]', '', tags)

    return tags

df['tags'] = df['tags'].apply(clean_tags)
```

## REMOVE DUPLICATE TAG WORDS
### Purpose: Avoid repetition like "music music song"


```python
def remove_duplicate_words(text):
    words = text.split()
    return ' '.join(set(words))

df['tags'] = df['tags'].apply(remove_duplicate_words)
```

## TAG COUNT FEATURE
### Purpose: Measure metadata richness


```python
df['tag_count'] = df['tags'].apply(lambda x: 0 if x == '[none]' else len(x.split('|')))
```

```python
df['has_description'] = df['description'].apply(
    lambda x: False if pd.isna(x) or x.strip() == '' else True
)
```

```python
df.drop(['description'], axis=1, inplace=True)
```

```python
df = df.dropna()
```

```python

# Rename for consistency
df.rename(columns={'view_count': 'views'}, inplace=True)

# Create engagement rate
df['engagement_rate'] = (df['likes'] + df['comment_count']) / df['views']

# Monetization proxy (very important)
df['monetization_score'] = df['views'] * df['engagement_rate']
```

### Category Mapping

```python
category_map={1:"Autos & Vehicles",2:"Music",10:"Comedy",15:"Science & Technology",17:"Anime/Animation",19:"Classics",20:"Comedy",22:"Drama"
,23:"Family",24:"Foreign",25:"Horror",26:"Sci-Fi/Fantasy",27:"Thriller",28:"Shorts",29:"Shows"
}
```

```python
df['category_name'] = df['categoryId'].map(category_map)
```

```python
def time_bucket(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

df['time_bucket'] = df['publish_hour'].apply(time_bucket)
```

```python
# STANDARDIZE COLUMN NAMES (SNAKE_CASE)

df.rename(columns={
    'publishedAt': 'published_at',
    'channelId': 'channel_id',
    'channelTitle': 'channel_title',
    'categoryId': 'category_id',
}, inplace=True)
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)
```

```python
# =========================================
# FINAL COLUMN ORDERING (DASHBOARD READY)
# =========================================

df = df[[
    # 🔹 1. IDENTIFIERS
    'video_id',
    'title',
    'channel_id',
    'channel_title',
    'category_id',
    'category_name',

    # 🔹 2. TIME (RAW + DERIVED)
    'published_at',
    'publish_time',
    'publish_date',
    'publish_hour',
    'time_bucket',
    'trending_date',
    'trending_year',
    'trending_month',
    'trending_day',

    # 🔹 3. PERFORMANCE (RAW METRICS)
    'views',
    'likes',
    'comment_count',
    'dislikes',

    # 🔹 4. ENGAGEMENT METRICS
    'likes_ratio',
    'comments_ratio',
    'engagement_rate',

    # 🔹 5. VIRALITY & GROWTH
    'days_to_trend',
    'trend_duration',

    # 🔹 6. MONETIZATION
    'monetization_score',

    # 🔹 7. CONTENT FEATURES
    'tags',
    'tag_count',
    'has_description',

    # 🔹 8. FLAGS
    'comments_disabled',
    'ratings_disabled'
]]
```
