## 🟩  DATA CLEANING & FEATURE TRANSFORMATION (Cell 19)

### Purpose
Handle outliers and prepare features for meaningful analysis  

### Why This Step
- Remove low-view noise (views < 1000)  
- Cap unrealistic ratios (engagement_rate, comments_per_1000_views)  
- Apply log transformation to handle skewness  

### Dashboard Connection
Improves accuracy of all visualizations, especially Overview & Timing dashboards  

### Insight
Stabilizes extreme values and makes patterns more interpretable  

```python
df = df[df['views'] >= 1000]
df['engagement_rate'] = df['engagement_rate'].clip(upper=1)
df['comments_per_1000_views'] = df['comments_per_1000_views'].clip(upper=50)
df['views_log'] = np.log1p(df['views'])
df['likes_log'] = np.log1p(df['likes'])
df['comments_log'] = np.log1p(df['comment_count'])
df['virality_score_log'] = np.log1p(df['virality_score'])
```

# Correlation Analysis

## 🟩  CORRELATION ANALYSIS

### Purpose
Examine relationships between key performance metrics such as views, likes, comments, engagement rate, and virality score  

### Why This Plot
A heatmap is used to visualize correlation strength and direction between variables in a compact form  

### Dashboard Connection
Supports multiple dashboards by validating relationships:
- Overview Dashboard → engagement funnel consistency  
- Category Dashboard → performance comparison logic  
- Channel Dashboard → engagement vs reach behavior  
- Timing & Virality Dashboard → virality relationships  

### Insight
- Views show strong positive correlation with likes and comments, confirming that higher reach drives higher interaction  
- Virality score is highly correlated with views, indicating fast-growing videos tend to achieve higher reach  
- Engagement rate has weak or near-zero correlation with views, suggesting that audience interaction quality is independent of scale  

### Key Takeaway
Different metrics capture different dimensions of performance — popularity (views) and engagement (interaction quality) should be analyzed separately  

```python
corr = df[['views_log','likes_log','comments_log','engagement_rate','virality_score_log']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

# Trend Speed Segmentation

## 🟩 TREND SPEED SEGMENTATION

### Purpose
Categorize videos based on how quickly they become trending after being published  

### Why This Step
Segmenting videos into groups (Instant, Fast, Moderate, Slow) helps simplify analysis of virality patterns  

### Dashboard Connection
Primarily supports:
- Timing & Virality Dashboard → time-to-trend analysis :contentReference[oaicite:0]{index=0}  
- India Dashboard → region-specific virality behavior :contentReference[oaicite:1]{index=1}  

### Insight
- A large proportion of videos fall into Instant and Moderate categories, indicating rapid content cycles  
- Slow trending videos are significantly fewer, showing delayed virality is uncommon  
- This distribution highlights that successful content typically gains traction quickly  

### Key Takeaway
Speed of growth is a critical factor in content success — early momentum strongly influences overall performance  

```python
df['trend_speed'] = pd.cut(df['days_to_trend'],
                          bins=[-1,1,3,7,100],
                          labels=['Instant','Fast','Moderate','Slow'])

sns.countplot(x='trend_speed', data=df)
plt.title("Trend Speed Distribution")
plt.show()
```
