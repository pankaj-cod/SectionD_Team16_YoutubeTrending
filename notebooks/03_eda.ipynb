#1. IMPORT + LOAD

### 1.1 IMPORT LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['font.family'] = 'DejaVu Sans'
```

### 1.2 LOAD DATA

```python
df = pd.read_csv('/content/drive/MyDrive/gen ai capstone/cleaned_india_youtube_data.csv')
```

```python
print("Shape:", df.shape)
df.head()
```

# 2. DATA VALIDATION

 ### 2.1 DATA QUALITY CHECK

```python
print("Missing Values:\n", df.isnull().sum())
print("Duplicate Rows:", df.duplicated().sum())
```

#3 Category Analysis





```python
plt.figure(figsize=(12,6))
df['category_name'].value_counts().plot(kind='bar')

plt.title("Trending Videos by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```

# 4 Performance Analysis

## 4.1 Views Distribution

```python
plt.figure(figsize=(10,5))
sns.histplot(df['views'], bins=50)

plt.title("Views Distribution")
plt.show()
```

## 4.2 Likes Distribution

```python
plt.figure(figsize=(10,5))
sns.histplot(df['likes'], bins=50)

plt.title("Likes Distribution")
plt.show()
```

## 4.3 Comments Distribution

```python
plt.figure(figsize=(10,5))
sns.histplot(df['comment_count'], bins=50)

plt.title("Comments Distribution")
plt.show()
```

# 5 Time Based Analysis

## 5.1 Publish Hour

```python
plt.figure(figsize=(12,6))

df.groupby('publish_hour')['views'].mean().plot()

plt.title("Publish Hour vs Average Views")
plt.xlabel("Publish Hour")
plt.ylabel("Average Views")

plt.show()
```

## 5.2 Time Bucket Analysis

```python
plt.figure(figsize=(10,5))

df.groupby('time_bucket')['views'].mean().plot(kind='bar')

plt.title("Time Bucket vs Views")
plt.show()
```

## 5.3 Trending Month

```python
plt.figure(figsize=(12,6))

df['trending_month'].value_counts().sort_index().plot(kind='bar')

plt.title("Trending Month Distribution")

plt.show()
```

## 5.4 Trending Day

```python
plt.figure(figsize=(12,6))

df['trending_day'].value_counts().plot(kind='bar')

plt.title("Trending Day Distribution")

plt.show()
```

# 6 Virality Analysis

##6.2 Trend Duration

```python
plt.figure(figsize=(10,5))

sns.histplot(df['trend_duration'], bins=30)

plt.title("Trend Duration Distribution")

plt.show()
```

## 8 Category vs Engagement

# 9 — Description Analysis

```python
plt.figure(figsize=(8,5))

sns.barplot(x='has_description', y='views', data=df)

plt.title("Description vs Views")

plt.show()
```

#10 Monetization Analysis

## 10.1 Monetization Score

```python
plt.figure(figsize=(10,5))

sns.histplot(df['monetization_score'], bins=50)

plt.title("Monetization Score Distribution")

plt.show()
```

##10.2 Monetization vs Category

```python
plt.figure(figsize=(12,6))

sns.boxplot(x='category_name', y='monetization_score', data=df)

plt.xticks(rotation=45)

plt.title("Category vs Monetization")

plt.show()
```

## 10.3 Monetization vs Views

```python
plt.figure(figsize=(10,5))

sns.scatterplot(x='views', y='monetization_score', data=df)

plt.title("Views vs Monetization Score")

plt.show()
```

##11 Channel Analysis

```python
top_channels = df.groupby('channel_title')['views'].sum().sort_values(ascending=False).head(20)

plt.figure(figsize=(12,6))

top_channels.plot(kind='bar')

plt.title("Top Channels by Views")

plt.show()
```
