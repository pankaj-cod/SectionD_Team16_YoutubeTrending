## 📂 Dataset Loading for Multiple Countries

### 🔹 Approach Used

The same code structure is reused for multiple countries by **only changing the file path** of the dataset. Each country has its own CSV file, and switching between datasets is done by updating the path accordingly.

```python
df = pd.read_csv("/content/drive/MyDrive/DVA_capstone/Raw_datasets/US_youtube_trending_data.csv")
```

---

### 🔹 Key Idea

- The logic remains **constant**
- Only the dataset path changes:
  - `US_youtube_trending_data.csv`
  - `IN_youtube_trending_data.csv`
  - `GB_youtube_trending_data.csv`
  - *etc.*

---

### ✅ Benefits

- **Code Reusability** → The same code works across all country datasets  
- **Consistency** → Ensures uniform data loading and structure  
- **Scalability** → Easily extendable to additional countries without modifying logic  

---

### 🎯 Summary

This approach simplifies dataset handling by separating **data location (file path)** from **processing logic**, making the workflow efficient, maintainable, and scalable for multi-country analysis.

## HANDLE MISSING VALUES
### Purpose: Ensure no null issues in analytics

````python
df['description'] = df['description'].fillna('')
df['tags'] = df['tags'].fillna('')
````

##DROP IRRELEVANT COLUMNS
### Purpose: Remove non-analytical fields

````python
df.drop(['thumbnail_link'], axis=1, inplace=True)
````

## DATATYPE CONVERSION
### Purpose: Enable time-based analysis

````python
# Fix datetime
df['publish_time'] = pd.to_datetime(df['publishedAt'])
df.drop(['publishedAt'], axis=1, inplace=True)
df['trending_date'] = pd.to_datetime(df['trending_date'])
````

````python
df['comments_disabled'] = df['comments_disabled'].astype(int)
df['ratings_disabled'] = df['ratings_disabled'].astype(int)
````

````python
df['categoryId'] = df['categoryId'].astype('int8')
df['comments_disabled'] = df['comments_disabled'].astype('int8')
df['ratings_disabled'] = df['ratings_disabled'].astype('int8')
````

## REMOVE INVALID RECORDS
### Purpose: Ensure data quality

````python
df = df[df['view_count'] > 0]
````

## REMOVE DUPLICATES
### Purpose: Avoid duplicate daily entries

````python
df = df.drop_duplicates(subset=['video_id', 'trending_date'])
````

````python
df.shape
````

##FEATURE ENGINEERING
### Purpose: Create analytical metrics

````python
# Engagement metrics
df['engagement_rate'] = (
    df['likes'] + df['dislikes'] + df['comment_count']
) / df['view_count']
df['likes_ratio'] = df['likes'] / (df['likes'] + df['dislikes'] + 1)
df['comments_per_1000_views'] = df['comment_count'] / (df['view_count'] + 1) * 1000
````

````python
df.info()
````

````python
# Time-based features
df['publish_hour'] = df['publish_time'].dt.hour
df['publish_day'] = df['publish_time'].dt.day_name()
````

````python
# Virality metric
df['days_to_trend'] = (
    df['trending_date']-
    df['publish_time']
).dt.days
df['days_to_trend'] = df['days_to_trend'].apply(lambda x: x if x >= 0 else 0)
df['virality_score'] = df['view_count'] / (df['days_to_trend'] + 1)
````

## Content Features
### Purpose: Analyse metadata and SEO strategy

````python
df['title_length'] = df['title'].apply(len)
df['tag_count'] = df['tags'].apply(lambda x: len(x.split('|')) if x != '[none]' else 0)
````

## TAG CLEANING - BASIC NORMALIZATION
### Purpose: Standardize tag format

````python
import re

def clean_tags(tags):
    if tags == '[none]' or tags == '':
        return ''

    tags = tags.lower()  # normalize case
    tags = tags.replace('|', ' ')  # replace separator

    # remove special characters
    tags = re.sub(r'[^a-zA-Z ]', '', tags)

    return tags

df['tags'] = df['tags'].apply(clean_tags)
````

## REMOVE DUPLICATE TAG WORDS
### Purpose: Avoid repetition like "music music song"


````python
def remove_duplicate_words(text):
    words = text.split()
    return ' '.join(set(words))

df['tags'] = df['tags'].apply(remove_duplicate_words)
````
